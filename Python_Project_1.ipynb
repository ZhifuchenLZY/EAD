```python
import pandas as pd
import numpy as np
df=pd.read_csv("/content/Popularity of Programming Languages from 2004 to 2024.csv")
```

```python
np.random.seed(42)

# Randomly set as NaN (5%)
percentage_null = 0.05
total_values = df.iloc[:, 1:].size
num_nulls = int(total_values * percentage_null)

random_indices = np.random.choice(df.index.size * (df.shape[1] - 1), num_nulls, replace=False)

for index in random_indices:
    row = index // (df.shape[1] - 1)
    col = index % (df.shape[1] - 1) + 1
    df.iat[row, col] = np.nan
```

```python
df.info()
```

```python
# Converting 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

print(df.dtypes)
```

```python
df.interpolate(method='linear', inplace=True)
```

```python
# Verify if null values have been handled
print(df.isnull().sum())
```

```python
# Filtering out languages with an average popularity less than 4%
language_columns = df.columns[1:]
popular_languages = [col for col in language_columns if df[col].mean() <= 4]
print(popular_languages)
```

```python
# line plot

import matplotlib.pyplot as plt
plt.figure(figsize=(14, 8))

for column in df.columns[1:]:
    plt.plot(df['Date'], df[column], label=column, alpha=0.8)

plt.xlabel('Date')
plt.ylabel('Popularity %')
plt.title('Popularity Trends of Programming Languages (2004-2024)')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title='Languages')
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt


language_columns = df.columns[1:]
popular_languages = [col for col in language_columns if df[col].mean() > 4]

df_popular = df[['Date'] + popular_languages]

plt.figure(figsize=(10, 6))
for column in df_popular.columns[1:]:
    plt.plot(df_popular['Date'], df_popular[column], label=column, alpha=0.8)

plt.xlabel('Date')
plt.ylabel('Popularity %')
plt.title('Trends of Popular Programming Languages (>4% Popularity)')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1), title='Languages')
plt.tight_layout()

plt.show()
```

```python
# heat map

import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = df_popular.iloc[:, 1:].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation Matrix of Programming Language Popularity Over Time")
plt.show()
```
