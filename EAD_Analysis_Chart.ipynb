```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# First Step :importing the library
%matplotlib inline
```

```python
# Scend load the data
data = pd.read_csv('processed_data.csv')
```

```python
import matplotlib.pyplot as plt

# Set the graph size
plt.figure(figsize=(14, 8))


# Iterate through each programming language column and draw a line graph
for language in data.columns[1:]:  # Ignore the 'Date' column and start traversing from the second column
    plt.plot(data['Date'], data[language], label=language)


# Setting up the chart title and labels
plt.title("Popularity Trends of Programming Languages (2004-2024)")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.legend(title="Language", loc='upper right', bbox_to_anchor=(1.3, 1))
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert the Date column to date format, specifying the format as “year-month-day”.
data['Date'] = pd.to_datetime(data['Date'], format='%Y-%m-%d')
```

```python
# Convert the Date column to date format
data['Date'] = pd.to_datetime(data['Date'])

# Find the latest date
latest_data = data[data['Date'] == data['Date'].max()]
latest_data = latest_data.drop(columns=['Date'])

# Get the top 10 most popular languages
top_10_languages = latest_data.T.nlargest(10, latest_data.index[0]).index
print("Top 10 languages:", top_10_languages)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

for language in top_10_languages:
    plt.plot(data['Date'], data[language], label=language)

plt.title("Popularity Trends of Top 10 Programming Languages (2004-2024)")
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.legend(title="Language", loc='upper right')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Select the most recent year's data and sort by popularity from highest to lowest
latest_year = data['Date'].max()
latest_data = data[data['Date'] == latest_year].drop(columns=['Date']).melt(var_name='Language', value_name='Popularity')
latest_data = latest_data.sort_values(by='Popularity', ascending=False)

plt.figure(figsize=(14, 8))

sns.barplot(x='Language', y='Popularity', data=latest_data, palette="viridis")

plt.title(f"Programming Languages Popularity in {latest_year.year}")
plt.xlabel("Programming Language")
plt.ylabel("Popularity")

# Resize and rotate the x-axis labels
plt.xticks(rotation=45, fontsize=8)

plt.show()

```

```python
import seaborn as sns

oldest_year = data['Date'].min()
oldest_data = data[data['Date'] == oldest_year].drop(columns=['Date']).melt(var_name='Language', value_name='Popularity')
oldest_data = oldest_data.sort_values(by='Popularity', ascending=False)

plt.figure(figsize=(14, 8))

sns.barplot(x='Language', y='Popularity', data=oldest_data, palette="viridis")

plt.title(f"Programming Languages Popularity in {oldest_year.year}")
plt.xlabel("Programming Language")
plt.ylabel("Popularity")

plt.xticks(rotation=45, fontsize=8)

plt.show()
```

```python
import seaborn as sns

latest_year = data['Date'].max()
latest_data = data[data['Date'] == latest_year].drop(columns=['Date'])

top_10_languages = latest_data.T.nlargest(10, latest_data.index[0]).index

# Extract the popularity data of these top 10 languages from the raw data
top_10_data = data[top_10_languages]

# Calculate the correlation matrix for the top 10 languages
correlation_matrix = top_10_data.corr()

# Plotting heat maps
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Correlation'})
plt.title("Correlation Heatmap of Top 10 Programming Languages")
plt.show()




```

```python
popularity_std = data.drop(columns=['Date']).std()

popularity_std = popularity_std.sort_values(ascending=False)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 8))
sns.barplot(x=popularity_std.index, y=popularity_std.values, palette="viridis")
plt.title("Standard Deviation of Programming Language Popularity (2004-2024)")
plt.xlabel("Programming Language")
plt.ylabel("Standard Deviation of Popularity")
plt.xticks(rotation=45, fontsize=8)
plt.show()


```

```python
relative_data = data.drop(columns=['Date'])
relative_data.set_index(data['Date'], inplace=True)
relative_data = relative_data.div(relative_data.sum(axis=1), axis=0)


import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))
relative_data.plot.area(alpha=0.6, cmap="tab20")
plt.title("Relative Popularity of Programming Languages Over Time (2004-2024)")
plt.xlabel("Year")
plt.ylabel("Relative Popularity (%)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=8)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

language_data = data.drop(columns=['Date']).T

scaler = StandardScaler()
language_data_scaled = scaler.fit_transform(language_data)

pca = PCA(n_components=2)
language_data_scaled_pca = pca.fit_transform(language_data_scaled)

pca_df = pd.DataFrame(language_data_scaled_pca, columns=['PC1', 'PC2'])
pca_df['Language'] = language_data.index

plt.figure(figsize=(12, 8))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', s=100, alpha=0.7)

for i in range(pca_df.shape[0]):
    plt.text(pca_df['PC1'][i], pca_df['PC2'][i], pca_df['Language'][i], fontsize=9)

plt.title("PCA of Programming Language Popularity Trends")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()

```
